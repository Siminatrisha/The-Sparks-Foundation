### The most affected countries by terrorist attack

```python
data['country_txt'].value_counts()
plt.figure(figsize=(20,12))
plt.pie(data['country_txt'].value_counts()[:20],labels=data['country_txt'].value_counts()[:20].index,autopct='%1.1f%%')
plt.show()
plt.figure(figsize=(20,8))
a=sns.countplot(data['country_txt'])
```

### Region most affected by terrorist attack

```python
plt.figure(figsize=(20,12))
sns.barplot(x=data['region_txt'].value_counts().index,y=data['region_txt'].value_counts().values)
plt.title('Regions affected by attacks',fontsize=20)
plt.xticks(rotation=45)
plt.xlabel('Regions',fontsize=20)
plt.ylabel('Number of attacks',fontsize=20)
plt.show()
```

##### Middle East&North Africa is the most attacked region

### Type of attacked used by a terrorist

```python
attack = data.loc[:,'attacktype1':'attacktype1_txt']
kill = data.loc[:,'nkill']
kill = pd.concat([attack, kill], axis=1)
kill_df= kill.pivot_table(columns='attacktype1_txt', values='nkill', aggfunc='sum')
kill_df
```

```python
labels = kill_df.columns.tolist() 
transpoze = kill_df.T # transpoze
values = transpoze.values.tolist()
fig, ax = plt.subplots(figsize=(17, 17), subplot_kw=dict(aspect="equal"))
plt.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.legend(labels, loc='upper right') # location legend
plt.show()
```

##### Armed Assaults are the most attacked terrorist type

### Data Visualization

```python
sns.heatmap(df.corr())
plt.show()
```

```python
df['Ship Mode'].value_counts()
```

```python
plt.title('Ship Mode')
plt.pie(df['Ship Mode'].value_counts(),labels=df['Ship Mode'].value_counts().index,autopct='%1.1f%%')
plt.show()
sns.countplot(df['Ship Mode'])
```

### Visualizing Segment

```python
df['Segment'].value_counts()
```

```python
sns.pairplot(df,hue='Segment')
```

### Visualizing Category

```python
df['Category'].value_counts()
```

```python
plt.title('Category')
plt.pie(df['Category'].value_counts(),labels=df['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
sns.countplot(df['Category'])
```

```python
plt.figure(figsize=(7,7))
plt.title('Sub-Category')
plt.pie(df['Sub-Category'].value_counts(),labels=df['Sub-Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
plt.figure(figsize=(15,8))
sns.countplot(df['Sub-Category'])
```

### Visualization of State

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(10,10))
a=df.groupby('State')['Profit'].sum()
a.plot.bar()
```

### Regional profit

```python
plt.title('Regional Profit')
plt.pie(df['Region'].value_counts(),labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
sns.countplot(df['Region'])
```

###  Profit vs Discount

```python
sns.lineplot(data=df,x='Discount',y='Profit',label='Profit')
plt.show()
```

###  Dependency of Sales,Profit,Discount

```python
plt.style.use('seaborn')
df.plot(kind='scatter',x='Sales',y='Profit',c='Discount',colormap='magma')
plt.ylabel('Profit')
plt.show()
```

###  Cities with minimum amount of profit

```python
cities=df[['City','Profit']]
min_cities=pd.DataFrame(cities.groupby('City')['Profit'].sum().reset_index().sort_values('Profit',ascending=True))
```
